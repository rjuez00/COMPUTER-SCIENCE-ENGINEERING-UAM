### Probar el error y la validación del Clasificador

como todavía no hemos implementado todo el clasificador no hemos puesto nada en el constructor, pero si tenemos el error y la validación implementada, así que simpularemos un clasificador generando una lista y modificandola

```python
datasetTest = dataset.extraeDatos(estrategiasimple.creaParticiones(dataset)[0].indicesTest)
```

#### Generamos la clasificación
Como no pertenece a esta práctica crear el clasificador hemos simulado que clasificabamos el dataset, hemos creado una lista con todo errores, otra con ningun error, y otra con algunos errores

```python
listaclasificada_sin_errores = datasetTest["Class"].tolist()
listaclasificada_todo_errores = [i+1 for i in datasetTest["Class"]]
listaclasificada_algunos_errores = [i+1 if i%2 == 0 else i for i in datasetTest["Class"]]
```

#### Creamos el clasificador
Lo creamos sin argumentos de momento porque no pertenece a esta práctica implementar la clase completa, solo el error y la validación

```python
clasificador=Clasificador.ClasificadorNaiveBayes()
```

#### Probamos el error

```python
print(f"Probar clasificación sin errores {clasificador.error(datasetTest, listaclasificada_sin_errores)}")
print(f"Probar clasificación con errores {clasificador.error(datasetTest, listaclasificada_todo_errores)}")
print(f"Probar clasificación con errores parciales {clasificador.error(datasetTest, listaclasificada_algunos_errores)}")
```

#### Probamos la validación

```python
clasificador.validacion(particionado = estrategiasimple, dataset = dataset, clasificador = clasificador)
```
