## 4. Predicción de fugas de clientes en una compañía telefónica (7 puntos)

Una compañía telefónica está interesada en que desarrollemos un modelo que prediga los **100 clientes actuales** (dataset de explotaación) que tienen más probabilidad de abandonar la compañía. Para ello nos proporcionan una base de datos **fuga_clientes_empresa_telefonica_construccion.csv** con casos etiquetados, que usaremos para construir nuestro modelo de predicción.

Los campos de esta base de datos son:

* **Customer ID**

* **network_age**: antigüedad del cliente en días

* **Customer tenure in months:** antigüedad del cliente en meses

* **Total Spend in Months 1 and 2:** gasto total del cliente en los meses de referencia 1 y 2

* **Total SMS Spend:** gasto total en SMS

* **Total Data Spend:** gasto total en datos/internet

* **Total Data Consumption:** consumo total de datos (en KB) durante el período de estudio

* **Total Unique Calls:** número total de llamadas únicas

* **Total Onnet spend:** gasto total en llamadas a otros usuarios de la misma red de telefonía

* **Total Offnet spend:** gasto total en llamadas a otros usuarios de redes de telefonía diferentes

* **Total Call centre complaint calls:** número de llamadas de queja al call center

* **Network type subscription in Month 1:** suscripción de tipo de red en el mes 1. Esto indica la suscripción de red preferida de un cliente, lo que puede indicar su tipo de dispositivo: servicio 2G o 3G

* **Network type subscription in Month 2:** igual que el anterior pero en el mes posterior

* **Churn Status**: el valor es 1 si el cliente abandona la compañía telefónica, 0 si permanece en ella

* **Most Loved Competitor network in Month 1:** qué otro proveedor de la competencia prefiere el cliente en el mes 1. En realidad es un conjunto de columnas, cada una enfocada en un proveedor particular

* **Most Loved Competitor network in Month 2:** qué otro proveedor de la competencia prefiere el cliente en el mes 2. En realidad es un conjunto de columnas, cada una enfocada en un proveedor particular

La variable a predecir es **Churn Status**: el valor es 1 si el cliente **abandona** la compañía, 0 si no la abandona.

La compañía también nos proporciona otra base de datos, **fuga_clientes_empresa_telefonica_explotacion.csv**, con información sobre clientes de los que no se sabe ahora mismo si van a permanecer o no en la compañía. Por tanto en esta segunda base de datos todos los valores de la columna **Churn Status** son missing values (NaN).

La compañía nos pide que proporcionemos los IDs de los 100 clientes de la base de datos de explotación que con mayor probabilidad vayan a abandonar la compañía. Para ello proporcionaremos como entregable a la compañía un archivo csv con una sola columna, **Customer ID**, y 100 filas que contengan los IDs de los clientes seleccionados.

El fichero **ejemplo_fichero_predicciones.csv** contiene un ejemplo con el formato solicitado para este archivo.

# What are we looking for?

We want to increase RECALL as much as possible because having a False Negatives is much worst than having a False Positive. We assumed this because in this case, the company would more likely prefer taking measures to make clients stay that are not going to leave, than not taking measures with the ones that are about to leave.
Due to this, we decided to look at F1 to balance, but taking F2 more into account, as it minimizes False Negatives without being as drastic as RECALL, accuracy was also taken into account.

For testing, at first, we approached the problem with a traditional train_test_split() but we found out that we could choose the scores cross_val_score() used so what we did was to use a 5 cross-validation fold to which we passed the scores we wanted to measure. We ended up using the function cross_validate(), which did all the tests for us.
As you can see in the results, AdaBoost was by far the best one achieving the highest accuracy, recall, F2, and F1 score, although decision_tree achieved high marks as well but even though accuracy was close we value a lot F2 score that is why we finally chose AdaBoost, we believe that AdaBoost improves on the algorithm of decision_trees that is why they both obtain the best results.

There are 2 sections, one for tinkering and achieving the best scores, and one for graphing which we used to check the hyperparameters of AdaBoost, knn, Logistic Regression (although we got always the same result for any value of C), and Decision Trees, we used those graphs to know where to start hyperparameter tuning, when we see the scores visually is easier to spot highlights.

# HyperParameter Tuning

### GaussianNB

```python
scoring = {'accuracy' : make_scorer(accuracy_score), 
           'precision' : make_scorer(precision_score),
           'recall' : make_scorer(recall_score), 
           'f1_score' : make_scorer(f1_score),
           'f2_score' : make_scorer(fbeta_score, beta=2)}
for key, value in cross_validate(GaussianNB(), X, y, cv=5, scoring=scoring).items():
    print(f"{key}: {value.mean()} +- {value.std()}")
```

### KNeighborsClassifier

```python

scoring = {'accuracy' : make_scorer(accuracy_score), 
           'precision' : make_scorer(precision_score),
           'recall' : make_scorer(recall_score), 
           'f1_score' : make_scorer(f1_score),
           'f2_score' : make_scorer(fbeta_score, beta=2)}
for key, value in cross_validate(KNeighborsClassifier(n_neighbors=1), X, y, cv=5, scoring=scoring).items():
    print(f"{key}: {value.mean()} +- {value.std()}")
```

### Decision Tree

```python
%matplotlib inline


# Decision Tree Graph
score_f2_mean = []
score_acc_mean = []
score_f2_variance_up = []
score_f2_variance_down = []
score_acc_variance_up = []
score_acc_variance_down = []


X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)


for i in range(1, 50):
    tempclf = DecisionTreeClassifier(max_depth=i)  
    testing = cross_validate(tempclf, X, y, cv=5, scoring=scoring)
    score_f2_mean.append(testing['test_accuracy'].mean())
    score_acc_mean.append(testing['test_f2_score'].mean())
    score_f2_variance_up.append(testing['test_f2_score'].mean() + testing['test_f2_score'].std())
    score_f2_variance_down.append(testing['test_f2_score'].mean() - testing['test_f2_score'].std())
    score_acc_variance_up.append(testing['test_accuracy'].mean() + testing['test_accuracy'].std())
    score_acc_variance_down.append(testing['test_accuracy'].mean() - testing['test_accuracy'].std())
    

plt.plot(range(1,50), score_f2_mean, label="f2")
plt.plot(range(1,50), score_acc_mean, label="accuracy")
plt.fill_between(range(1,50), score_f2_variance_up, score_f2_variance_down, color="orange", alpha=0.2)
plt.fill_between(range(1,50), score_acc_variance_up, score_acc_variance_down, color="blue", alpha=0.2)
plt.legend()

fig = plt.gcf()
fig.set_size_inches(17.5, 10.5)

plt.title('F2 and accuracy scores for decision Tree')
plt.xlabel('Depth')
plt.ylabel('Scores')
plt.show()
```
